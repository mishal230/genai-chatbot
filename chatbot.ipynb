```python
# Use a pipeline as a high-level helper
from transformers import pipeline

# Initialize the text generation pipeline with the DialoGPT model
pipe = pipeline("text-generation", model="microsoft/DialoGPT-medium")

# Define the initial user input
user_input = "Who are you?"

# Generate a response from the model
response = pipe(user_input, max_length=100, num_return_sequences=1)

# Get the generated text (the model's response)
bot_response = response[0]['generated_text']

# Print the response
print("Bot:", bot_response)
```

```python
# Initialize an empty conversation history
conversation_history = ""

# Define a loop for interaction
while True:
    user_input = input("You: ")

    # Append the user input to the conversation history
    conversation_history += f"User: {user_input}\n"

    # Generate a response based on the entire conversation history
    response = pipe(conversation_history, max_length=100, num_return_sequences=1)

    # Extract the bot's reply
    bot_response = response[0]['generated_text'].split("Bot:")[-1].strip()

    # Print the response and append to history
    print("Bot:", bot_response)
    conversation_history += f"Bot: {bot_response}\n"
```
